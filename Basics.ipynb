As we start to work through this data, we'll check in and take stock of our insights and learnings.

#### Key callouts
1. Students in more popular majors don't necessarily make more money.
2. Additionally, students who majored in female majority majors also didn't tend to make more money.
    
In both cases (1 and 2), larger populations also meant lower median salaries. This makes sense since more general majors were probably easier and thus, yielded lower salaries overall.There isnt a clear link between employees and median salary other than there aren't any majors for which there a can be many participants AND high median salary amount.
